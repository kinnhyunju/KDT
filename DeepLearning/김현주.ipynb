7. 이진분류 모델 구현을 간략하게 코드 작성하세요
    - 피처 5개, 클래스 2개, 층 4개

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchmetrics.classification import BinaryF1Score

class custom_model(nn.Module):
    def __init__(self) :
        super().__init__()
        self.input_layer = nn.Linear(5,10)
        self.hidden_layer1 = nn.Linear(10,8)
        self.hidden_layer2 = nn.Linear(8,4)
        self.output_layer = nn.Linear(4,1)
    
    def forward(self, x):
        y = self.input_layer(x)
        y = F.leaky_relu(y)
        y = self.hidden_layer1(y)
        y = F.leaky_relu(y)
        y = self.hidden_layer2(y)
        y = F.leaky_relu(y)
        return self.output_layer(y)
```
