##### 준공실적

```python
import pandas as pd
```

```python
file = r'C:\Users\kdp\KDT_06\KDT\EX_PY06\KDT\Project02\주택건설_준공실적_월계__20240718021412.xlsx'
```

```python
# XLSX ==> DataFrame 형태 저장
complete1 = pd.read_excel(file)
complete1
```

```python
# 0번행, 1번 컬럼 삭제
complete1=complete1.drop(0)
complete1=complete1.drop('부문명(1)',axis=1)
complete1
```

```python
# 총계 부분만 추출
complete2 = complete1.drop([2,3,4,5,6])
complete2
```

```python
# 행열 전치
complete = complete2.transpose()
complete
```

```python
# 0번열 컬럼 설정 / 0번행 삭제
complete = complete.rename(columns=complete.iloc[0])
complete = complete.drop('구  분(1)')
complete
```

```python
# 연도별로 데이터 나누기
complete_18 = complete.iloc[0:12]
complete_19 = complete.iloc[12:24]
complete_20 = complete.iloc[24:36]
complete_21 = complete.iloc[36:48]
complete_22 = complete.iloc[48:60]
```

```python
# 데이터프레임 합치기 위해 행인덱스 통일, 열 인덱스는 연도로 나누기
month = ['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월']
complete_18.columns=[2018]
complete_18.index = month
complete_19.columns=[2019]
complete_19.index = month
complete_20.columns=[2020]
complete_20.index = month
complete_21.columns=[2021]
complete_21.index = month
complete_22.columns=[2022]
complete_22.index = month
```

```python
# 데이터 프레임 합치기 concat() 함수 사용
sum_complete = pd.concat([complete_18,complete_19,complete_20,complete_21,complete_22],axis=1)
```

```python
# 준공실적 평균 행 추가
sum_complete.loc['평균'] = sum_complete.mean()
sum_complete
```

##### 주택 매매가격 변동률

```python
file = r'C:\Users\kdp\KDT_06\KDT\EX_PY06\KDT\Project02\2018~2022 주택매매가격변동률.xlsx'
```

```python
# 엑셀파일 ==> DataFrame으로 저장
houseDF1 = pd.read_excel(file)
houseDF1
```

```python
# 행열 전치
houseDF=houseDF1.transpose()
houseDF = houseDF.drop('행정구역별(1)')
houseDF.columns = ['변동률']
houseDF.head(5)
```

```python
# 연도별로 데이터 나누기
houseDF_18 = houseDF.iloc[0:12]
houseDF_19 = houseDF.iloc[12:24]
houseDF_20 = houseDF.iloc[24:36]
houseDF_21 = houseDF.iloc[36:48]
houseDF_22 = houseDF.iloc[48:60]
```

```python
# 데이터프레임 합치기 위해 행인덱스 통일, 열 인덱스는 연도로 나누기
month = ['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월']
houseDF_18.columns=['2018']
houseDF_18.index = month
houseDF_19.columns=['2019']
houseDF_19.index = month
houseDF_20.columns=['2020']
houseDF_20.index = month
houseDF_21.columns=['2021']
houseDF_21.index = month
houseDF_22.columns=['2022']
houseDF_22.index = month
```

```python
# 데이터 프레임 합치기 concat() 함수 사용
sum_houseDF = pd.concat([houseDF_18,houseDF_19,houseDF_20,houseDF_21,houseDF_22],axis=1)
```

```python
# 주택 매매가 변동률 평균 행 추가
sum_houseDF.loc['평균'] = [sum_houseDF['2018'].mean(),sum_houseDF['2019'].mean(),sum_houseDF['2020'].mean(),sum_houseDF['2021'].mean(),sum_houseDF['2022'].mean()]
sum_houseDF
```

##### 준공실적과 매매가격 변동률 비교

```python
# 값 숫자형식으로 변환
sum_complete = sum_complete.astype('Float64')
sum_houseDF = sum_houseDF.astype('float64')
sum_complete.info()
```
