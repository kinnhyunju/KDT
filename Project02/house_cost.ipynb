##### 매매 가격 지수

```python
import pandas as pd
```

```python
file_name = 'house_sale.xlsx'
```

```python
# 엑셀파일 ==> DataFrame으로 저장
sale = pd.read_excel(file_name)
sale
```

```python
# 행열 전치
housesale = sale.transpose()
housesale
```

```python
# 0번열 컬럼 설정 / 0번행, 1번행 데이터 삭제
housesale = housesale.rename(columns=housesale.iloc[0])
housesale= housesale.drop(['주택유형별(1)','지역별(1)'])
housesale
```

```python
# 연도별로 데이터 나누기
sale_18 = housesale.iloc[0:12]
sale_19 = housesale.iloc[12:24]
sale_20 = housesale.iloc[24:36]
sale_21 = housesale.iloc[36:48]
sale_22 = housesale.iloc[48:60]
```

```python
# 종합 칼럼만 추출
sum_sale_18 = sale_18.drop(['아파트','연립다세대','단독주택'],axis=1)
sum_sale_19 = sale_19.drop(['아파트','연립다세대','단독주택'],axis=1)
sum_sale_20 = sale_20.drop(['아파트','연립다세대','단독주택'],axis=1)
sum_sale_21 = sale_21.drop(['아파트','연립다세대','단독주택'],axis=1)
sum_sale_22 = sale_22.drop(['아파트','연립다세대','단독주택'],axis=1)
```

```python
# 데이터프레임 합치기 위해 행인덱스 통일, 열 인덱스는 연도로 나누기
month = ['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월']
sum_sale_18.columns=[2018]
sum_sale_18.index = month
sum_sale_19.columns=[2019]
sum_sale_19.index = month
sum_sale_20.columns=[2020]
sum_sale_20.index = month
sum_sale_21.columns=[2021]
sum_sale_21.index = month
sum_sale_22.columns=[2022]
sum_sale_22.index = month
```

```python
# 데이터 프레임 합치기 concat() 함수 사용
sum_sale = pd.concat([sum_sale_18,sum_sale_19,sum_sale_20,sum_sale_21,sum_sale_22],axis=1)
```

```python
sum_sale
```

```python
'''
상품의 가격이 수요가 있지만 공급이 적을 때 꾸준히 오른다고 가정하면
주택 매매 가격은 2018년 1월부터 2022년 2월까지 꾸준히 상승하고 있지만 
'''
```

```python
# 매매가격지수 평균 행 추가
sum_sale.loc['평균'] = sum_sale.mean()
sum_sale
```
