    * 기본확인 파악된 내용들 처리

```python
# [3-1] 불필요 데이터 처리
copyDF = dataDF.drop(index=0)
print(f'확인\n{copyDF.head(3)}')
```

```python
# [3-2] 결측치 데이터 처리 => isna() / isnull()
copyDF.isna().sum()
```

```python
# - '1984','1985','1986' 컬럼에서 결측치 데이터만 추출
a = copyDF[['1984','1985','1986']].isna()
a[a.sum(axis=1)>0]
```

```python
# - '전출지별' 
copyDF['전출지별'].ffill(inplace=True)

# - 수치데이터 결측치 처리
# pd.set_option('future.no_silent_downcasting', True)
copyDF = copyDF.fillna(0)

# 데이터의 일부를 추출해서 값을 변환할 때 inplace=True 잘 살펴보기!
```

```python
# [3-3] 이상치 데이터 처리
copyDF.replace('-',0,inplace=True)
```

```python
copyDF.shape
```

```python
# [3-4] 자료형 변환 : 48개 컬럼을 int로 형 변환
# 딕셔너리 형태로 {1970:'int',1971:'int',...,2017:'int',} 만들기
d_ = dict(zip(copyDF.columns[2:].tolist(), ['int']*(copyDF.shape[1]-2)))

# copyDF = copyDF.astype('int')를 하면 컬럼명도 숫자로 변환됨(?)
copyDF = copyDF.astype(d_)
```

- [4] 데이터 가공
    - 대구에서 전지역 전입 인구수 분석
    - 전출지별 - '대구광역시'
    - 전입지별 - '대구광역시','전국' 제외

```python
# - 대구 광역시 데이터만 추출
mask = (copyDF['전출지별']=='대구광역시')
cityDF = copyDF[mask]
cityDF
```

```python
# - 전출지별 컬럼 삭제
cityDF = cityDF.drop('전출지별',axis=1)

# - 전입지별 컬럼 행 인덱스 설정
cityDF.set_index('전입지별',inplace=True)
```

```python
# - 대구광역시, 전국 행 삭제
cityDF.drop(['전국','대구광역시'],inplace=True)
```

```python
# - 가공데이터 확인
print(cityDF.dtypes, cityDF.columns, cityDF.index,sep='\n\n')
```

- [5] 분석 : 대구에서 전지역 전입 인구수 분석

```python
# - [5-1] 대구 & 타지역
cityDF['합계'] = cityDF.sum(axis=1)
cityDF
```

```python
# 한글 폰트 설정
from matplotlib import font_manager as fm
from matplotlib import rc
font_file = r'C:\Windows\fonts\malgunsl.ttf'

# 폰트 이름 가져오기
font_name = fm.FontProperties(fname = font_file).get_name()

# 폰트패밀리 이름 지정
rc('font',family = font_name)
```

```python
# x축 : 서울특별시 ~ 세종특별자치시
# y축 : 인구수 총합
bar_colors = ['tab:blue','tab:orange','tab:green','tab:red','tab:purple','tab:brown','tab:pink','tab:gray','tab:olive','tab:cyan']
plt.figure(figsize=(10,5))
plt.bar(cityDF.index[:7],cityDF['합계'][:7],color = bar_colors, width=0.6)

plt.title('[ 대구광역시 전출 인구 - 광역시별 ]')
plt.xlabel('< 전입지 >')
plt.ylabel('이동 인구수')
plt.show()
```

```python
# x축 : 경기도 ~ 제주특별자치도
# y축 : 인구수 총합
colors = ['tab:blue','tab:orange','tab:green','tab:red','tab:purple','tab:brown','tab:pink','tab:gray','tab:olive','tab:cyan']
plt.figure(figsize=(10,7))
plt.bar(cityDF.index[7:].to_list(),cityDF['합계'][7:], color = bar_colors, width=0.6)

plt.title('[ 대구광역시 전출 인구 - 자치도별 ]')
plt.xlabel('< 전입지 >')
plt.ylabel('이동 인구수')
plt.show()
```
