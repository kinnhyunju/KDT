- [3] 데이터 전처리
    * 기본확인 파악된 내용들 처리

```python
# [3-1] 불필요 데이터 처리
copyDF = dataDF.drop(index=0)
print(f'확인\n{copyDF.head(3)}')
```

```python
# [3-2] 결측치 데이터 처리 => isna() / isnull()
copyDF.isna().sum()
```

```python
# - '1984','1985','1986' 컬럼에서 결측치 데이터만 추출
a = copyDF[['1984','1985','1986']].isna()
a[a.sum(axis=1)>0]
```

```python
# - '전출지별' 
copyDF['전출지별'].ffill(inplace=True)

# - 수치데이터 결측치 처리
# pd.set_option('future.no_silent_downcasting', True)
copyDF = copyDF.fillna(0)

# 데이터의 일부를 추출해서 값을 변환할 때 inplace=True 잘 살펴보기!
```

```python
# [3-3] 이상치 데이터 처리
copyDF.replace('-',0,inplace=True)
```

```python
copyDF.shape
```

```python
# [3-4] 자료형 변환 : 48개 컬럼을 int로 형 변환
# 딕셔너리 형태로 {1970:'int',1971:'int',...,2017:'int',} 만들기
d_ = dict(zip(copyDF.columns[2:].tolist(), ['int']*(copyDF.shape[1]-2)))

# copyDF = copyDF.astype('int')를 하면 컬럼명도 숫자로 변환됨(?)
copyDF = copyDF.astype(d_)
```

- [4] 데이터 가공
    - 대구에서 전지역 전입 인구수 분석
    - 전출지별 - '대구광역시'
    - 전입지별 - '대구광역시','전국' 제외

```python
# - 대구 광역시 데이터만 추출
mask = (copyDF['전출지별']=='대구광역시')
cityDF = copyDF[mask]
cityDF
```

```python
# - 전출지별 컬럼 삭제
cityDF = cityDF.drop('전출지별',axis=1)

# - 전입지별 컬럼 행 인덱스 설정
cityDF.set_index('전입지별',inplace=True)
```

```python
# - 대구광역시, 전국 행 삭제
cityDF.drop(['전국','대구광역시'],inplace=True)
```

```python
# - 가공데이터 확인
print(cityDF.dtypes, cityDF.columns, cityDF.index,sep='\n\n')
```

- [5] 히스토그램으로 특정 컬럼 데이터 분포 시각화

```python
# - 2000년도 데이터의 분포
cityDF.columns
```

```python
# 값 기준 정렬 ==> [기본]오름차순
cityDF['2000'].sort_values(ascending=False)
```

```python
# 2000년도 대구 지역에서 타지역으로 이동 인구수 분포
plt.hist(cityDF['2000'],bins=15,histtype='stepfilled')
plt.hist(cityDF['2017'],bins=15,histtype='step')
plt.show()
```

```python
n,bin,obj = plt.hist(cityDF['2000'],bins=15,histtype='stepfilled')
# n : 구간마다 들어있는 값 개수
# bin : 구간 (기준)
# obj : 그래프 개체
print(n)
print(bin.tolist())
print(obj)
labels = list(map(int,bin.tolist()))
plt.xticks(bin[::2],labels[::2],rotation = 45)
plt.show()
```

```python
# 2000년도 대구 지역에서 타지역으로 이동 인구수 분포
fig = plt.figure(figsize=(10,5))
ax1,ax2 = fig.subplots(1,2,sharey=True)
ax1.hist(cityDF['2000'],bins=15,histtype='stepfilled')
ax2.hist(cityDF['2017'],bins=15,histtype='step')
plt.show()
```
