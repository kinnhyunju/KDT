# titanic 자료 분석

```python
# [1] 모듈 로딩
import pandas as pd
```

```python
# [2] 데이터 불러오기
file = 'titanic.csv'
```

```python
# [3] CSV파일 ==> DataFrame형태로 변환
titDF = pd.read_csv(file)
titDF
```

```python
titDF.isna().sum()
# age : 177 / embarked : 2 / deck : 688 / embark_town : 2
```

```python
# 겹치는 항목 제거
df1 = titDF.drop(['deck','sex','adult_male','alive','class','embarked'],axis=1)
```

```python
df1.dropna(inplace=True)
```
