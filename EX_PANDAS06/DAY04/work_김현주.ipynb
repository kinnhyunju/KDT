# titanic 자료 분석

```python
# [1] 모듈 로딩
import pandas as pd
```

```python
# [2] 데이터 불러오기
file = 'titanic.csv'
```

```python
# [3] CSV파일 ==> DataFrame형태로 변환
titDF = pd.read_csv(file)
titDF
```

```python
# [4] 데이터 확인
# - 기본정보 확인
titDF.info()
```

```python
# - 컬럼과 실제 데이터 타입 확인
print(titDF.head(), titDF.tail(),sep='\n\n')
```

```python
# - 컬럼별 데이터 분포 확인
titDF.describe()
```

```python
# [5] 데이터 전처리
# - [5-1] 결측치 체크
titDF.isna().sum()
```

```python
# 중복 컬럼 삭제
titDF.drop(['deck','sex','adult_male','alive','class','embarked'],inplace=True,axis=1)
```

```python
# - 설정값 : 특정 컬럼만 NaN여부 검사 진행 후 결정 : subset
titDF.dropna(axis=0, subset=['embark_town'],inplace=True)
titDF.isna().sum()
```

```python
titDF['embark_town'].value_counts()
```

```python
# 탑승 지역별 탑승 인원
'''
Southampton    644
Cherbourg      168
Queenstown      77 '''
```

#### 탑승 지역별 객실 1등급 비율
##### Southampton : 19.7%
##### Cherbourg   : 50.6%
##### Queenstown  :  2.6%

```python
# Southampton 중 1등석 탑승 인원 127명
titDF[(titDF['embark_town']=='Southampton')&(titDF['pclass']==1)]
```

```python
# Cherbourg 중 1등석 탑승 인원 85명
titDF[(titDF['embark_town']=='Cherbourg')&(titDF['pclass']==1)]
```

```python
# Queenstown 중 1등석 탑승 인원 2명
titDF[(titDF['embark_town']=='Queenstown')&(titDF['pclass']==1)]
```
