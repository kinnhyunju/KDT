#### [파이 그래프]
- 컬럼(변수)의 데이터 분포 비율을 시각화

```python
# [1] 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# [2] 데이터 준비
F_DATA = r'C:\Users\kdp\KDT_06\KDT\EX_PY06\KDT\EX_PANDAS06\DAY04\auto_mpg.csv'
```

```python
# [3] CSV ==> DataFrame으로 저장
dataDF = pd.read_csv(F_DATA)
```

```python
# - 컬럼별 타입 확인
dataDF.dtypes
```

```python
# [4] 특정 컬럼의 데이터 비율
# - 자동차 생산 국가에 대한 비율
dataDF['origin'].astype('category')
```

```python
# - 자동차 생산 국가에 대한 비율
round(dataDF['origin'].value_counts()/dataDF['origin'].count(),2)
```

```python
originSR = dataDF['origin'].value_counts()
x = originSR.index.to_list()
h = originSR.values.tolist()
x,h
```

```python
# - 자동차 생산 국가에 대한 비율 시각화 - bar 그래프
originBar = plt.bar(x,h, color = ['tab:cyan','tab:purple', 'tab:pink'], edgecolor = ['black'])
plt.xticks([1,2,3],['US','EU','JP'])
plt.bar_label(originBar,h, label_type='center')
plt.show()
```

```python
# - 자동차 생산 국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['US','EU','JP'],autopct='%.1f%%')
plt.show()
```

```python
# - 자동차 생산 국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['US','EU','JP'],autopct='%.1f%%',
        counterclock= False, startangle=90)
plt.show()
```

```python
# - 자동차 생산 국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['US','EU','JP'],autopct='%.1f%%',
        counterclock= False, startangle=90,explode=[0,0,0.1])
plt.show()
```

```python
# - 자동차 생산 국가에 대한 비율 시각화 - pie 그래프
plt.pie(h,labels=['US','EU','JP'],autopct='%.1f%%',
        counterclock= False, startangle=90,wedgeprops={'linewidth':2, 'width':0.72,'edgecolor':'plum'})
plt.show()
```
