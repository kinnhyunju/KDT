#### [Matplotlib 활용한 시각화]
- matplotlib.pyplot 모듈을 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재

```python
# [1] 모듈 로딩
# 패키지명.모듈명
import matplotlib.pyplot as plt
```

```python
# [2] 그래프 그리기 - 선(Line) 그래프
# plot(y축 데이터)
plt.plot([1,2,3,4])
plt.xlabel('DATA')
plt.ylabel('RESULT')
plt.show()
# 마지막에 들어가는 명령어를 출력
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터)
plt.plot([10,20,30],[5,10,15])
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터, fmf 포맷스트링)
# fmf (포맷스트링) : 그래프 색상, 두께, 스타일 등 지정
plt.plot([10,20,30],[5,10,15],'r--')
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터, fmf 포맷스트링)
# fmf (포맷스트링) : 그래프 색상, 두께, 스타일 등 지정
plt.plot([10,20,30],[5,10,15],'b>--')
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plot(x축 데이터, y축 데이터, line2D 속성으로 설정)
# - color='red', marker='o', linestyle='dashed'
# - markersize=10, linewidth=4
# - marker edge color => mec = 'blue'
# - marker edge width => mew = 5
plt.plot([10,20,30],[5,10,15],
         color='red', marker='o', linestyle='dashed',
         markersize=10, linewidth=4,
          mec = 'blue',mew = 5)
plt.show()
```

```python
# [2] 그래프 그리기 - Line(선) 그래프
# plt.plot([10,20,30],[15,25,35],'bo-')
# plt.plot([10,20,30],[18,29,25],'ro--')
# plt.plot([10,20,30],[25,25,25],'k^-')

# 한번에 여러개 그리기
plt.plot([10,20,30],[15,25,35],'bo-',[10,20,30],[18,29,25],'ro--',[10,20,30],[25,25,25],'k^-')
plt.show()
```

```python
# [3] 그래프에 부가적인 기능 추가
# - 그래프 제목
# - X축, Y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 label 속성 값을 범례에서 사용함
plt.plot([10,20,30],[15,25,20],'bo--',label = 'V')
plt.plot([10,20,30],[5,30,28],'ro--',label = '2')
plt.title('[Matplotlib Line 2D]')
plt.xlabel('Data')
plt.ylabel('Result')
plt.legend()
plt.show()
```

```python
# [3] 그래프에 부가적인 기능 추가
# - 그래프 제목
# - X축, Y축 라벨
# - 여러개 그래프 경우 범례 => 그래프의 label 속성 값을 범례에서 사용함
#                            loc 속성으로 위치 지정 가능 [기:best]
plt.plot([10,20,30],[15,25,20],'bo--',label = 'V')
plt.plot([10,20,30],[5,30,28],'ro--',label = '2')
plt.title('[Matplotlib Line 2D]')
plt.xlabel('Data')
plt.ylabel('Result')
plt.legend(loc = 'lower right')
plt.show()
```
