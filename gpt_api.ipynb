```python
import pandas as pd
store_df = pd.read_csv('store.csv')
stores = store_df['매장명'][7640:].to_list()
stores[:3]
```

```python
import openai

# 매장명과 음식 종류 개수 정의
# stores = ['맥도날드', '스타벅스', '파리바게트', '처갓집 양념치킨', '버거킹', '엔젤리너스', '신전 떡볶이', '베스킨 라빈스', '달인의 찜닭']
food_category_counts = [1, 2, 3]

def generate_gpt_answer(store, count):
    # 프롬프트 작성
    prompt = (
        f"""
        이제부터 너는 음식을 주문하는 사람이 되어 한 문단으로 이루어진 완전한 자연어 주문 텍스트를 생성하는 역할을 할거야.
        주문 텍스트의 내용에는 매장명, 음식명, 수량이 포함되어야 해. 그리고 마침표나 느낌표 같은 특수문자는 모두 제거해줘.
        그리고 주문 텍스트의 시작 부분에 반드시 인사말을 넣을 필요는 없어. 넣어도 되고 넣지 않아도 되고 바리에이션을 줘.
        {store} 매장명을 이용하여 {count}개의 음식 종류를 넣고, 각 음식의 수량은 자유롭게 정한 뒤에 주문 텍스트를 생성해줘.
        """
    )

    # API 호출
    response = openai.chat.completions.create(
        model = "gpt-4o-mini", # 모델 선택
        messages = [{"role" : "system", "content" : "너는 창의적인 텍스트 생성 도우미야."}, # 시스템에게 역할 부여
                    {"role" : "user", "content" : prompt}], # 사용자는 prompt 객체를 시스템에게 전달
        temperature = 1.0, # 창의성 정도
        max_tokens = 200 # 응답 길이 제한
    )

    return response

for i in food_category_counts:
    for j in stores:
        response = generate_gpt_answer(j, i)
        result_text = response.choices[0].message.content

        with open('gpt_text.txt', mode='a') as f:
            f.write(result_text + '\n')
```
